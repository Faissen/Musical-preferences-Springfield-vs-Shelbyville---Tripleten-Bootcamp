## 3.2 Check duplicates

```python
duplicates = music_df.duplicated().sum()
print(f'The number of duplicate rows in the dataset is: {duplicates}')  
```

Remove the duplicated lines and then check the null values.

```python
#Remove the duplicates and reset the index (drop=True remove the prevous indexes)
music_df = music_df.drop_duplicates().reset_index(drop=True)  
```

```python
#Check intrinsic duplicates in each column
#for col in music_df.columns:
#    print(f'Unique values in {col}: {music_df[col].unique()}')

print(music_df['genre'].unique())
```

```python
#Normalize text
music_df['genre'] = (
    music_df['genre']
        .str.strip()
        .str.lower()
        .str.normalize('NFKD') # Normalize unicode characters - base letter + accents
        .str.encode('ascii', errors='ignore') # Remove accents by ignoring non-ascii characters, text becomes bytes
        .str.decode('utf-8') # Decode back to string
)

print(np.sort(music_df['genre'].dropna().unique())) # Ordered alphabetically
```

There are some variants of the same genre: hip, hop, hiphop, hip-hop. Create a function to replace all variants to hiphop.

```python
def replace_genre(df, column, wrong_genres, correct_genre):
    for wrong_genre in wrong_genres:
        df[column] = df[column].replace(wrong_genre, correct_genre)
    return df

replace_genre(music_df,'genre', ['hip hop', 'hiphop', 'hip'], 'hip-hop')
```

# 3.4 Data types

Change the following data types:
* City may be a category instead of object (only 2 possibilities)
* time should be time instead of object
* Day may be a category instead of object (only 3 possibilities)

```python
music_df['city'] = music_df['city'].astype('category')
music_df['time'] = pd.to_datetime(music_df['time'], format='%H:%M:%S')
music_df['day'] = music_df['day'].astype('category')

print(music_df.dtypes)
```
